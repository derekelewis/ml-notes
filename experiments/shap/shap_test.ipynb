```python
import transformers
import shap
import torch

pmodel = transformers.pipeline(
    "text-classification",
    "microsoft/phi-2",
    trust_remote_code=True,
    device_map="mps",
    torch_dtype=torch.bfloat16,
)

data = ["Alice: Who is the current president of the United States?\nBob:"]

pmodel.model.config.task_specific_params = dict(
    {"text-generation": {"max_new_tokens": 20}}
)

pmodel.model.config.is_decoder = True
exp = shap.Explainer(pmodel, pmodel.tokenizer)
shap_values = exp(data, max_evals=15)
shap.plots.text(shap_values)
```
